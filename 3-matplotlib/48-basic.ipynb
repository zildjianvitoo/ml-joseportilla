```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = x * 2
y
```

```python
plt.plot(x,y,)
plt.title("Relasi antara x dan Y")
plt.xlabel("X axis judul")
plt.ylabel("Y axis judul")
plt.xlim(0,6)
plt.ylim(0,15)
plt.savefig(fname="kocak.png", )
```
