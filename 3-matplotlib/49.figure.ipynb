```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,11)
y = x * 2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.show()
```

```python
axes0x = np.arange(0,11)
axes0y = np.sin(axes0x)

axes1x = np.arange(0,11)
axes1y = axes1x ** 2
```

```python
fig = plt.figure()
axes0 = fig.add_axes([0,0,0.5,0.5])
axes0.plot(axes0x,axes0y)

axes1 = fig.add_axes([0.6,0,0.5,0.5])

axes1.plot(axes1x,axes1y)

axes2 = fig.add_axes([1.2, 0, 0.5, 0.5])

axes2.plot(axes1x, axes1y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([1.1,0.45,0.3,0.3])


a = np.arange(0,11)
b = a ** 4
axes1.plot(a, b)
axes1.set_title("Title for axes 1")
axes1.set_xlabel("Base number")
axes1.set_ylabel("Times 2 from base number")


axes2.plot(a,b)
axes2.set_title("Zoom")
axes2.set_xlim(2,6)
axes2.set_ylim(0,5000)
axes2.set_ylabel("")
```

```python
a = np.linspace(0,10,11)

b = a ** 4

figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b)

plt.show()
```

```python
fig = plt.figure(dpi=100, figsize=(6, 3), facecolor="0.9")
axes = fig.add_axes([0,0,1,1])

axes.plot(a,b);

fig.savefig("Kocak banget.png",bbox_inches="tight")
```
