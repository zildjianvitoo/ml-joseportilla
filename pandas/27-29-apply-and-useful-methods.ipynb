```python
def last_four(num):
    return str(num)[-4:]
```

```python
last_four(df["CC Number"][0])
```

```python
df["Last Four Number"] = df["CC Number"].apply(last_four)
df.head()
```

```python
def how_expensive_therestaurant(total_bill):
    if total_bill < 10:
        return "$"
    elif total_bill >= 10 and total_bill < 30:
        return "$$"
    else:
        return "$$$"
```

```python
df["Expensive"] = df["total_bill"].apply(how_expensive_therestaurant)
df.head()
```

```python
df["total_bill"].apply(lambda total_bill: total_bill * 2)
```

```python
def quality_of_tip(total_bill,tip):
    if tip / total_bill > 0.25:
        return "Generous"
    else:
        return "Kemurahan"
```

```python
df = df.drop("Quality of Tip",axis=1)
df.head()
```

```python
df["Quality of Tip"] = df[["total_bill","tip"]].apply(lambda df: quality_of_tip(df["total_bill"],df["tip"]),axis=1)
```

```python
df["Quality of Tip"] = np.vectorize(quality_of_tip)(df["total_bill"],df["tip"])
```
