```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    "C:\\Coding\\python-projects\\ml-jose-portilla\\jose\\05-Seaborn\\dm_office_sales.csv"
)
df.head()
```

```python
plt.figure(figsize=(12,5))

sns.countplot(data=df, x="level of education",hue="division",palette="Set2")
```

```python
df = pd.read_csv(
    "C:\\Coding\\python-projects\\ml-jose-portilla\\jose\\05-Seaborn\\StudentsPerformance.csv"
)

df.head()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(
    data=df,
    x="writing score",
    y="parental level of education",
    hue="gender",
    palette="Set2"
)
```

```python
plt.figure(figsize=(10, 10))
sns.violinplot(data=df, x="math score", y="parental level of education",hue="parental level of education", palette="Set2",bw_adjust=0.1)
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=df,x="gender",y="math score",hue="test preparation course",palette="Set2",s=5)

plt.legend(bbox_to_anchor=(1,0.1))
```

```python
plt.figure(figsize=(10,6))
sns.boxenplot(data=df,x="reading score", y="parental level of education",hue="gender",palette="Set2")
```
