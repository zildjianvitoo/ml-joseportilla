```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    "C:\\Coding\\python-projects\\ml-jose-portilla\\jose\\05-Seaborn\\dm_office_sales.csv"
)
df.head()
```

```python
plt.figure(figsize=(8,4))
sns.rugplot(data=df,x="salary",height=0.7)
```

```python
sns.displot(data=df,x="salary",bins=20,color="green")
```

```python
sns.histplot(data=df,x="salary",bins=20,color="green")
```

```python
sns.displot(data=df,x="salary",kind="hist")
```

```python
np.random.seed(42)

sample_ages = np.random.randint(0,100,200)
sample_ages
```

```python
sample_ages_df = pd.DataFrame(data=sample_ages,columns=["age"])
sample_ages_df.head()
```

```python
sns.displot(data=sample_ages_df,x="age",kind="kde")
```

```python
sns.kdeplot(data=sample_ages_df,x="age",clip=[0,100],color="green",bw_adjust=0.1,fill=True)
```

```python
sns.displot(data=sample_ages_df,x="age",bins=20)
```
