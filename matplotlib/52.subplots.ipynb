```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
a = np.linspace(10,20,11)
```

```python
y = x ** 3
b = np.cos(a)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
# fig.set_figwidth(10)
# fig.set_figheight(5)

fig.suptitle("Learn matplotlib",fontsize=16)

axes[0,0].set_title("Relasi antara x dan y")
axes[0,0].plot(x, y)
axes[0,0].set_xlabel("X")
axes[0,0].set_ylabel("Y")

axes[0, 1].plot(x, y)

axes[1,0].plot(a, b)

axes[1,1].set_title("Relasi antara a dan b")
axes[1,1].plot(a, b)
axes[1,1].set_xlabel("A")
axes[1,1].set_ylabel("B")

# fig.subplots_adjust(wspace=0.3,hspace=0.4)
fig.tight_layout()
fig.savefig("2 dimension axes figure.png",bbox_inches="tight")
```
