```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = x * 2
```

```python
fig,ax= plt.subplots()

ax.plot(x,x**2,color="green",lw=1,ls="--")
ax.plot(x,x**3)
ax.legend(labels=["x ** 2","x ** 3"])
```

```python
x2 = np.arange(0,11)

fig,ax = plt.subplots(figsize=(10,6))

line1, = ax.plot(x2,x2**2,color="green",lw=1,marker="s",markersize=10,markerfacecolor="red",markeredgewidth=5)
line2, = ax.plot(x2,x2 **2.5,color="purple",marker="P",ms=7)
ax.legend(labels=["x vs x^2","x vs x^2.5"])
```
