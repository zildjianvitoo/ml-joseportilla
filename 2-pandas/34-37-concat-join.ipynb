```python
registrations = pd.DataFrame(
    {"reg_id": [1, 2, 3, 4], "name": ["Andrew", "Bobo", "Claire", "David"]}
)
logins = pd.DataFrame(
    {"log_id": [1, 2, 3, 4], "name": ["Xavier", "Andrew", "Yolanda", "Bobo"]}
)
```

```python
merge_dataframe = pd.merge(logins,registrations,how="left",on="name")

def fill_with_random_value(value):
    if pd.isna(value) :
        return np.random.randint(1,10)
    else:
        return value

merge_dataframe["reg_id"] = merge_dataframe["reg_id"].apply(fill_with_random_value)
merge_dataframe
```

```python
pd.merge(left=registrations,right=logins,how="right",on="name")
```

```python
mergedf = pd.merge(logins, registrations, how="outer", on="name")

mergedf = mergedf.sort_values(by="name",ascending=False)
mergedf
```
