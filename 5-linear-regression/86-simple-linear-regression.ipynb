```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    "C:\\Coding\\python-projects\\ml-jose-portilla\\jose\\08-Linear-Regression-Models\\Advertising.csv"
)
```

```python
df.head()
```

```python
df["total_ads"] = df["TV"] + df["newspaper"] + df["radio"]
df.head()
```

```python
sns.scatterplot(data=df,x="total_ads",y="sales")
```

```python
sns.regplot(data=df,x="total_ads",y="sales")
```

```python
X = df["total_ads"]
y = df["sales"]
```

```python
b1,b0 = np.polyfit(X,y,deg=1)
```

```python
potential_spend = np.linspace(0,500,100)
```

```python
predicted_sales = b1 * potential_spend + b0
```

```python
sns.scatterplot(data=df,x="total_ads",y="sales")
plt.plot(potential_spend,predicted_sales,color="red")
```

```python
b3,b2,b1,b0 = np.polyfit(X,y,3)
```

```python
potential_spend = np.linspace(0,500,100)
```

```python
predicted_sales = b3*(potential_spend**3) + b2*(potential_spend**2) + b1*potential_spend + b0
```

```python
sns.scatterplot(data=df, x="total_ads", y="sales")
plt.plot(potential_spend,predicted_sales,color="red")
```
